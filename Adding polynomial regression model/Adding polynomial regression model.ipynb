```python
#importing libraries
import pandas as pd
import numpy as np
import datetime as dt
from datetime import datetime,date,time
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv('C:\\Users\\Harshita\\Downloads\\state_wise_daily.csv')
```

```python
df.head()
```

```python
df['Confirmed']=df['TT']
```

```python
df.isnull().sum()
```

```python
columns = ['Date','Date_YMD','Status','Confirmed']
```

```python
df = df[columns]
```

```python
df
```

```python
df=df[df.Status=='Confirmed']
```

```python
df
```

```python
df = df.drop('Status',axis=1)
```

```python
df
```

```python
df['Date']=pd.to_datetime(df['Date'])
df['Date']=df['Date'].map(dt.datetime.toordinal)
```

```python
df
```

```python
x=df[['Date']]
y=df[['Confirmed']]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
xtrain
```

```python
ytrain
```

## Linear Regression model

```python
from sklearn.linear_model import LinearRegression
lg = LinearRegression()
model = lg.fit(xtrain,ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
score = lg.score(xtest,ytest)
```

```python
score
```

```python
from sklearn.metrics import mean_squared_error
rmse=mean_squared_error(ytest,ypred)
```

```python
rmse
```

## Polynomial Regression model

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=20)
xpoly = poly_reg.fit_transform(xtrain)
pol_reg = LinearRegression()
model1 = pol_reg.fit(xpoly,ytrain)
xpolyt = poly_reg.fit_transform(xtest)
pred = pol_reg.predict(xpolyt)
from sklearn.metrics import mean_squared_error
rmse1=mean_squared_error(ytest,pred)
rmse1
```

```python
score1 = pol_reg.score(xpolyt,ytest)
```

```python
score1
```

- In Polynomial Regression when we increase our Degree it reduce the Root mean Squared error and improve our score also in compare to Linear Regression but we cannot increase the Degree of Polynomial to much because it increase the variance.So the Polynomial Regression model is ready to run.
